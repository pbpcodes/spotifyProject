```python
mostPopularSongs = dfTrack.query('popularity>90',inplace=False).sort_values('popularity',ascending=False)
mostPopularSongs.head()
```

```python
dfTrack.set_index('release_date',inplace=True)
dfTrack.index = pd.to_datetime(dfTrack.index)
dfTrack.head()
```

```python
dfTrack[['artists']].iloc[10]
```

```python
dfTrack['duration'] = dfTrack['duration_ms'].apply(lambda i:round(i/1000))
dfTrack.drop('duration_ms',inplace=True,axis=1)
```

```python
dfTrack.head()
```

```python
dfTrack.duration.head()
```

# Correlation - Pearson Method

```python
correlationDf = dfTrack.drop(['key','explicit','mode'],axis=1).corr(method='pearson')
```

```python
mpl.figure(figsize=(14,6))
```

```python
heatMap = sbn.heatmap(correlationDf,annot=True,fmt='.1g',vmax=1,vmin=-1,center=0,cmap='Greens', linewidths=1,linecolor='Black')
heatMap.set_title("Correlation Heatmap between Variables")
heatMap.set_xticklabels(heatMap.get_xticklabels(),rotation=90)
```

```python
sampleDf = dfTrack.sample(int(0.005*len(dfTrack)))
```

```python
sampleDf.head()
```

```python
mpl.figure(figsize=(10,6))
sbn.regplot(data=sampleDf,y="loudness",x="energy" ).set(title="Energy VS Loudness Correlation")
```

```python
mpl.figure(figsize=(10,6))
sbn.regplot(data=sampleDf,y="popularity",x="acousticness" ).set(title="Acousticness VS Popularity Correlation")
```

```python
dfTrack['dates']=dfTrack.index.get_level_values('release_date')
dfTrack.dates=pd.to_datetime(dfTrack.dates)
years = dfTrack.dates.dt.year
```

```python
sbn.displot(years,discrete=True,height=5,aspect=2,kind='hist').set(title="Number of songs per year since 1922")
```

```python
totalDuration = dfTrack.duration
figureDimensions = (20,8)
fig,axis = mpl.subplots(figsize=figureDimensions)
fig = sbn.barplot(x=years,y=totalDuration,ax=axis,errwidth=False).set(title="Duration of songs over the years")
mpl.xticks(rotation=90)
```

```python
dfFeatureGenre = pd.read_csv('SpotifyFeatures.csv')
```

```python
dfFeatureGenre
```

```python
mpl.title("Duration of songs in different genres")
sbn.color_palette("inferno",as_cmap=True)
sbn.barplot(x='duration_ms',y='genre',data=dfFeatureGenre)
mpl.xlabel("Duration in ms")
```
